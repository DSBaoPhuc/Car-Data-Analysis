# Import Libraries

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

import plotly.graph_objs as go
import plotly.express as px
```

# Loading Dataset

```python
data = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
data.head()
```

## Make new column name_2 contain the name of the car in dataset

```python
data["name_2"] = data.name.apply(lambda x : ' '.join(x.split(' ')[:1]))
data['name_2']
```

# Data Visualization

## Number of cars of each type and visualize with a bar chart

```python
data.name_2.value_counts()
```

```python
sns.countplot(data=data,x="name_2",palette="CMRmap")
plt.xticks(rotation=90)
plt.xlabel("Name",fontsize=10,color="black")
plt.ylabel("Name",fontsize=10,color="black")
plt.title("NAME COUNT",color="black")
plt.show()
```

### Use pie charts to visualize car data most common use

```python
labels = data["name_2"][:30].value_counts().index
sizes = data["name_2"][:30].value_counts()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99',"pink","yellow"]
plt.figure(figsize = (8,8))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors,shadow=True, startangle=45)
plt.title('name',color = 'red',fontsize = 15)
plt.show()
```

As can we see from the chart, the most common car model is the Maruti.

## Number of year and visualize it

```python
data.year.value_counts()
```

### Visualize number of each year with bar chart

```python
sns.countplot(data=data,x="year",palette="icefire")
plt.xticks(rotation=90)
plt.xlabel("YEAR",fontsize=10,color="RED")
plt.ylabel("COUNT",fontsize=10,color="RED")
plt.title("YEAR COUNT",color="RED")
plt.show()
```

So the highest count in 2017 with 466 instances, next 2015 (421), 2012 (415), 2013 (386), and 2014 (367).

### Visualize number of each year with pie chart

```python
labels = data["year"][:40].value_counts().index
sizes = data["year"][:40].value_counts()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99',"pink","yellow"]
plt.figure(figsize = (8,8))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors,shadow=True, startangle=45)
plt.title('Year',color = 'red',fontsize = 15)
plt.show()
```

The pie chart will visually represent the distribution of occurrences of the top 40 years from the year column, with each slice showing the proportion of each year. This helps in understanding the relative frequency of each year in a visually intuitive manner.

## Sunburst plot to visualize the distribution of these entries by year and Car name
